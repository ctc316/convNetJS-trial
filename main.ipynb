#### Obeserve and plot the loss over time

```python
iters = [500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000]
losses_1 = [0.0139068, 0.009719, 0.008254, 0.0074679, 0.006795, 0.006501, 0.006094, 0.0057255, 0.005443, 0.005448]

plt.plot(iters, losses_1)
plt.xticks(iters)
plt.show()
```

#### Halve the learning rate every 1000 iterations

```python
losses_2 = [0.013410, 0.009508, 0.005884, 0.005508, 0.004376, 0.004136, 0.003739, 0.003651, 0.003421, 0.003400]

plt.plot(iters, losses_1, label='Fixed Learing Rate')
plt.plot(iters, losses_2, label='Halved Learing Rate')
plt.xticks(iters)
plt.legend()
plt.show()
```
